## 3. Discussion


Signals constitute an important part of our daily life. Anything that carries some information is called a signal. A signal is defined as a single-valued function of one or more independent variables which contain some information. A signal is also defined as a physical quantity that varies with time, space or any other independent variable. A signal may be represented in time domain or frequency domain. Human speech is a familiar example of a signal. Electric current and voltage are also examples of signals. A signal can be a function of one or more independent variables. A signal may be a function of time, temperature, position, pressure, distance etc. If a signal depends on only one independent variable, it is called a one- dimensional signal, and if a signal depends on two independent variables, it is called a two- dimensional signal.

When we process a signal, this signal may undergo several manipulations involving the independent variable or the amplitude of the signal. The basic operations on sequences are as follows:
1. Time shifting
2. Time reversal
3. Time scaling
4. Amplitude scaling
5. Signal addition
6. Signal multiplication

The first three operations correspond to transformation in independent variable n of a signal. The last three operations correspond to transformation on amplitude of a signal.

## 5. Directions:

1. Using NumPy libraries, create a sinusoidal signal using the following formula:

> $x_1(t) = A \sin(2\pi f t + \phi)$

- where A represents the amplitude of the signal, f denotes the frequency of the signal and ϕ indicates the phase of the signal.

- Set A = 3.2, f = 2.4, and $\phi = \dfrac{3\pi}{4}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 3.2
f = 2.4
phi = 3 * np.pi / 4

# Time vector
t = np.linspace(0, 3, 1000)

# Generate sinusoidal signal
x1 = A * np.sin(2 * np.pi * f * t + phi)

# Plot
plt.figure(figsize=(10, 4))
plt.plot(t, x1)
plt.title(r'Sinusoidal Signal: $x_1(t) = A\sin(2\pi f t + \phi)$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

print(f'Amplitude: {A}, Frequency: {f} Hz, Phase: {np.degrees(phi):.1f} degrees')
print(f'Signal length: {len(x1)} samples over {t[-1]:.1f} seconds')
```


2. Using the created sinusoidal signal, perform time advance with 3 units. Then graph and compare the result signal to original sinusoidal signal.


```python
# Time advance: x1(t + 3) shifts the signal to the LEFT by 3 units
# To achieve this, we subtract 3 from the time variable
t_shifted = t - 3
x1_advance = A * np.sin(2 * np.pi * f * t_shifted + phi)

# Plot comparison
plt.figure(figsize=(10, 4))
plt.plot(t, x1, label='Original $x_1(t)$')
plt.plot(t, x1_advance, '--', label='Time Advance $x_1(t+3)$')
plt.title('Time Advance by 3 Units')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print('Time advance shifts the signal to the left.')
print('The advanced signal reaches its peaks/valleys earlier than the original.')
```

3. Perform time reversal with the created sinusoidal signal. Then graph and compare the result signal to original sinusoidal signal.

```python
# Time reversal: x1(-t) flips the signal about the vertical axis
x1_reversed = A * np.sin(2 * np.pi * f * (-t) + phi)

# Plot comparison
plt.figure(figsize=(10, 4))
plt.plot(t, x1, label='Original $x_1(t)$')
plt.plot(t, x1_reversed, '--', label='Time Reversal $x_1(-t)$')
plt.title('Time Reversal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print('Time reversal mirrors the signal across the vertical (amplitude) axis.')
print('Events that occurred early in the original now appear late, and vice versa.')
```

4. Perform time scaling using a=0.4 with the created sinusoidal signal. Then graph and compare the result signal to original sinusoidal signal.

```python
# Time scaling: x1(a*t) with a = 0.4 compresses/stretches the signal
# Since a < 1, the signal is stretched (period increases)
a_ts = 0.4
x1_scaled = A * np.sin(2 * np.pi * f * (a_ts * t) + phi)

# Plot comparison
plt.figure(figsize=(10, 4))
plt.plot(t, x1, label='Original $x_1(t)$')
plt.plot(t, x1_scaled, '--', label=f'Time Scaled $x_1({a_ts}t)$')
plt.title(f'Time Scaling with a = {a_ts}')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f'Time scaling by a = {a_ts} (< 1) stretches the signal.')
print('The effective frequency becomes f*a = {:.2f} Hz, so the period increases.'.format(f * a_ts))
```

5. Perform amplitude scaling using a=1/4 with the created sinusoidal signal. Then graph and compare the result signal to original sinusoidal signal.

```python
# Amplitude scaling: a * x1(t) with a = 1/4 reduces the signal's amplitude
a_amp = 1/4
x1_amp_scaled = a_amp * x1

# Plot comparison
plt.figure(figsize=(10, 4))
plt.plot(t, x1, label='Original $x_1(t)$')
plt.plot(t, x1_amp_scaled, '--', label=f'Amplitude Scaled $\frac{{1}}{{4}}x_1(t)$')
plt.title('Amplitude Scaling with a = 1/4')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f'Original amplitude: {A}')
print(f'Scaled amplitude: {A * a_amp}')
print('Amplitude scaling only affects the vertical magnitude; frequency and phase remain unchanged.')
```

6. Create a separate exponential signal using the following formula:
> $x_2(t) = a^t$

- Graph the result signal where a = 1.8

```python
# Exponential signal: x2(t) = a^t with a = 1.8
a_exp = 1.8
x2 = a_exp ** t

# Plot
plt.figure(figsize=(10, 4))
plt.plot(t, x2)
plt.title(r'Exponential Signal: $x_2(t) = 1.8^t$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

print(f'Base a = {a_exp} (> 1), so the signal grows exponentially over time.')
print(f'At t=0, x2(0) = {a_exp**0}, at t=1, x2(1) = {a_exp**1:.1f}, at t=3, x2(3) = {a_exp**3:.2f}')
```

7. Perform signal addition for x1 and x2 and plot the result.

```python
# Signal addition: x_sum(t) = x1(t) + x2(t)
# Both signals must have the same time vector
x_sum = x1 + x2

# Plot all three signals
plt.figure(figsize=(10, 6))
plt.plot(t, x1, label='$x_1(t)$ (sinusoidal)')
plt.plot(t, x2, label='$x_2(t)$ (exponential)')
plt.plot(t, x_sum, 'k-', linewidth=2, label='$x_{sum}(t) = x_1(t) + x_2(t)$')
plt.title('Signal Addition')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print('The sum combines the oscillating sinusoid with the growing exponential.')
print('The exponential component dominates at larger t values.')
```

8.	Perform signal multiplication for x1 and x2. Then graph and compare the product signal to summed signal.

```python
# Signal multiplication: x_prod(t) = x1(t) * x2(t)
x_prod = x1 * x2

# Plot comparison: product vs sum
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(t, x_sum, 'b-', label='$x_{sum}(t) = x_1(t) + x_2(t)$')
axes[0].set_title('Signal Addition')
axes[0].set_xlabel('Time (t)')
axes[0].set_ylabel('Amplitude')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(t, x_prod, 'r-', label='$x_{prod}(t) = x_1(t) \cdot x_2(t)$')
axes[1].set_title('Signal Multiplication')
axes[1].set_xlabel('Time (t)')
axes[1].set_ylabel('Amplitude')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

print('In multiplication, the exponential envelope modulates the sinusoidal amplitude.')
print('The sinusoidal oscillations grow in amplitude as t increases due to the exponential factor.')
print('In addition, the sinusoid rides on top of the exponential curve.')
```
